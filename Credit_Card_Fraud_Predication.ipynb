#OVER-SAMPLING

```python
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler

cc_df['normAmount'] =  StandardScaler().fit_transform(np.array(cc_df['Amount']).reshape(-1,1))
```

```python
cc_df = cc_df.drop(['Time', 'Amount'], axis=1)
```

```python
X , y = cc_df.drop('Class', axis=1) , cc_df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 
```

```python
smot = SMOTE()

X_train_sm , y_train_sm = smot.fit_resample(X_train, y_train)
```

```python
X_train_sm.shape, y_train_sm.shape
```

```python
y_train_sm.value_counts()
```

```python
logreg = LogisticRegression()
logreg.fit(X_train_sm, y_train_sm)
```

```python
y_ = logreg.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

print("Accuracy Score :", accuracy_score(y_test, y_))
print(classification_report(y_test, y_))
```
